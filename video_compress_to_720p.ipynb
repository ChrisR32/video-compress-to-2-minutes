```python
import subprocess
from pathlib import Path

# -----------------------------
# settings
# -----------------------------
input_video = "Comp Vision video.mp4"
output_video = "Group-121_Video.mp4"

# lower CRF = better quality/larger file
# higher CRF = lower quality/smaller file
# good range: 23 to 32
quality = 28

# -----------------------------
# compress video
# similar idea to HandBrake "Very Fast 720p30"
# -----------------------------
cmd = [
    "ffmpeg",
    "-i", input_video,

    # resize to 720p max height, keep aspect ratio
    "-vf", "scale=-2:720,fps=30",

    # video compression
    "-c:v", "libx264",
    "-preset", "veryfast",
    "-crf", str(quality),

    # audio compression
    "-c:a", "aac",
    "-b:a", "128k",

    # better mp4 compatibility
    "-movflags", "+faststart",

    output_video
]

subprocess.run(cmd, check=True)

print("compressed video saved as:", output_video)
```
